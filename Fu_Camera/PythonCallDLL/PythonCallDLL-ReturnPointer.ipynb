```python
from ctypes import *
import ctypes as ct
from numpy.ctypeslib import ndpointer
import matplotlib.pyplot as plt
```

```python
mydll = cdll.LoadLibrary("C:\\Users\\Agarwal\\Desktop\\CameraGithub\\Fu_Camera\\CameraDLL_ReturnPointer\\x64\\Debug\\CameraDLL_Fu.dll")
```

```python
mydll.sum(5,30)
```

```python
mydll.OpenCamera()  
```

```python
temp = 30.0

mydll.SetTemp(c_double(temp))
```

```python
exposureTime = 10.0
mydll.SetExposureTime(c_double(exposureTime))
```

```python
mydll.SetROIMod.argtypes = (ct.c_int, ct.c_int, ct.c_int, ct.c_int)
mydll.SetROIMod.restype = None
```

```python
hOffset = 0
vOffset = 0
width = 2048
hight = 2048
mydll.SetROIMod(hOffset, vOffset, width, hight)
```

```python
#mydll.CaptureSingleFrame.restype = ct.c_ushort * 10
mydll.CaptureSingleFrame.argtype = ct.c_int
mydll.CaptureSingleFrame.restype = ndpointer(dtype=ct.c_ushort, shape=(hight, width))
```

```python
imageData = mydll.CaptureSingleFrame(2)
```

```python
fig, ax = plt.subplots()
im = ax.imshow(imageData)

plt.show()
```

```python
mydll.BufRelease()
```
